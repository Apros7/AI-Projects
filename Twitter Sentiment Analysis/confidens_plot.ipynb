```python
import matplotlib.pyplot as plt
import matplotlib
import statistics
from math import sqrt
```

```python
names = ['Simple model', 'Afinn model', 'Tf-idf', 'Fine tuned bert']
x_values = [1, 2, 3, 4]
results = [0.61, 0.64, 0.82, 0.87]
n = 120000
```

```python
for i in range(4):
    p = results[i]
    x = x_values[i]
    confidence = 1.96 * sqrt((p*(1-p))/n)
    bottom = p-confidence
    top = p+confidence
    color='#2187bb'
    horizontal_line_width=0.25
    left = x - horizontal_line_width / 2
    right = x + horizontal_line_width / 2
    plt.xticks(x_values, names)
    plt.plot([x, x], [top, bottom], color=color)
    plt.plot([left, right], [top, top], color=color)
    plt.plot([left, right], [bottom, bottom], color=color)
    plt.plot(x, results[i], 'o', color='#f44336')
    plt.title("Model accuracy on test set")
    plt.xlabel("Model")
    plt.ylabel("Accuracy (%)")
```
